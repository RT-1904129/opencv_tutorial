here we do edge detection

```python
cap=cv2.VideoCapture(0)
def sketch(image):
    img_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)#gray operation
    img_gray_blur=cv2.GaussianBlur(img_gray,(5,5),0)#Bluring operation for filtering by kernal 5x5 matrix
    canny_edge=cv2.Canny(img_gray_blur,10,70)#Canny operation
    ret,mask=cv2.threshold(canny_edge,70,255,cv2.THRESH_BINARY)#threshold converting binery images of gray images
    return mask

while True:
    ret,frame=cap.read()
    cv2.imshow("our Live Sketch",sketch(frame))
    if cv2.waitKey(1)==13:
        break
cap.release()
cv2.destroyAllWindows()
```

Color filtering by webcam

we use hsv for range of colur range H=color range S=saturation V=intensity  

blue color range is from 110-130 so it is value of H

```python
import numpy as np
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    lower_range=np.array([110,50,50])
    upper_range=np.array([130,255,255])
    mask=cv2.inRange(hsv,lower_range,upper_range)#here we make matrix of under bound of frame matrix 
    cv2.imshow("show",mask)
    cv2.imshow("show1",frame)
    if cv2.waitKey(1)==13:
        break
cap.release()
cv2.destroyAllWindows()
```
